```python
# Import Pandas
import pandas as pd

# Display all columns
pd.set_option('display.max_columns', None)

# Display rows
pd.set_option('display.min_rows', 6)

# Read CSV files
gdp_data = pd.read_csv('gdp_data.csv')
country_codes = pd.read_csv('country_codes.csv')

# Merge data
merged_data = pd.merge(gdp_data, country_codes, on='country_code')

# Display the first 10 rows of dataset
print(merged_data.head(10))
```

```python
import matplotlib.pyplot as plt

# Group data by year and sum GDP column
plt.plot(merged_data.groupby('year')['value'].sum())

plt.xlabel('Year')
plt.ylabel('Global GDP')
plt.title('Time Series of Global GDP')
plt.show()
```

```python
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

# Suppress warnings
warnings.filterwarnings('ignore', category=FutureWarning) #suppressing warnings for my environment in particular

sns.boxplot(x='region', y='value', data=merged_data)
plt.xticks(rotation=45)
plt.title('GDP by Region')
plt.ylabel('GDP')
plt.show()
```

```python
sns.catplot(x='region', y='value', hue='income_group', kind='violin', data=merged_data)
plt.xticks(rotation=45)
plt.title('Interaction between GDP and Income Group by Region')
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_geo(merged_data,
                     locations="country_code",
                     size="value",
                     animation_frame="year",
                     projection="natural earth",
                     title="Yearly Global GDP by Country over time")
fig.show()
```

```python
import plotly.express as px

# Filter data for 2020
data_2020 = merged_data[merged_data['year'] == 2020]

fig = px.sunburst(data_2020,
                  path=['region', 'country_name'],
                  values='value',
                  title="GDP Distribution by Region and Country for the Year 2020")

fig.show()
```
